```python
df.drop(columns=['Survived'],inplace=True)
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean()) #clean missing values
```

```python
features=df[['Fare','Age']]
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)
```

```python
predicted=model.fit_predict(features)
```

```python
predicted=pd.DataFrame( predicted, columns=['Predicted Labels']  )
predicted.sample(5)
```

```python
predicted['Predicted Labels'].value_counts()
```

```python
temp=df[['Name','Age','Gender','Pclass','Fare']]
temp
```

```python
ans=pd.concat([temp,predicted],axis=1)
ans
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='Age',y='Fare',data=ans,hue='Predicted Labels',palette='dark')
```
